```julia
# Created by K.Nakamura on 2015-11-21.
# e-mail: nakamura-keita-kn@ynu.jp


typealias Quad4 Element{2,4}
typealias Quad8 Element{2,8}


function ShapeFunction(::Type{Quad4})
    function N(point)                          ##### Where does this "point" come from?
        A = Matrix{Float64}(1,4)
        ξ, η = point
        A[1] = 0.25 * (1-ξ) * (1-η)
        A[2] = 0.25 * (1+ξ) * (1-η)
        A[3] = 0.25 * (1+ξ) * (1+η)
        A[4] = 0.25 * (1-ξ) * (1+η)
        return A
    end
    function ∇N(point)
        A = Matrix{Float64}(2,4)
        ξ, η = point
        A[1] = 0.25 * (-1+η)
        A[2] = 0.25 * (-1+ξ)
        A[3] = 0.25 * ( 1-η)
        A[4] = 0.25 * (-1-ξ)
        A[5] = 0.25 * ( 1+η)
        A[6] = 0.25 * ( 1+ξ)
        A[7] = 0.25 * (-1-η)
        A[8] = 0.25 * ( 1-ξ)
        return A
    end
    return ShapeFunction(N,∇N)
end

function ShapeFunction(::Type{Quad8})
    function N(point)
        A = Matrix{Float64}(1,8)
        ξ, η = point
        A[1] = 0.25 * (1-ξ) * (1-η) * (-1-ξ-η)
        A[2] = 0.25 * (1+ξ) * (1-η) * (-1+ξ-η)
        A[3] = 0.25 * (1+ξ) * (1+η) * (-1+ξ+η)
        A[4] = 0.25 * (1-ξ) * (1+η) * (-1-ξ+η)
        A[5] = 0.5 * (1-ξ^2) * (1-η)
        A[6] = 0.5 * (1+ξ) * (1-η^2)
        A[7] = 0.5 * (1-ξ^2) * (1+η)
        A[8] = 0.5 * (1-ξ) * (1-η^2)
        return A
    end
    function ∇N(point)
        A = Matrix{Float64}(2,8)
        ξ, η = point
        A[1]  = 0.25 * (1-η) * (2ξ+η)
        A[2]  = 0.25 * (1-ξ) * (2η+ξ)
        A[3]  = 0.25 * (1-η) * (2ξ-η)
        A[4]  = 0.25 * (1+ξ) * (2η-ξ)
        A[5]  = 0.25 * (1+η) * (2ξ+η)
        A[6]  = 0.25 * (1+ξ) * (2η+ξ)
        A[7]  = 0.25 * (1+η) * (2ξ-η)
        A[8]  = 0.25 * (1-ξ) * (2η-ξ)
        A[9]  = -ξ * (1-η)
        A[10] = -0.5 * (1-ξ^2)
        A[11] =  0.5 * (1-η^2)
        A[12] = -η * (1+ξ)
        A[13] = -ξ * (1+η)
        A[14] =  0.5 * (1-ξ^2)
        A[15] = -0.5 * (1-η^2)
        A[16] = -η * (1-ξ)
        return A
    end
    return ShapeFunction(N,∇N)
end


function create_gauss_points(::Type{Quad4})
    gauss_line = GaussLine(2,2)
    gauss_points = [GaussPoint(i...) for i in gauss_line]  ####?? How does this comment work?
    return gauss_points
end

function create_gauss_points(::Type{Quad8})
    gauss_line = GaussLine(3,2)
    gauss_points = [GaussPoint(i...) for i in gauss_line]  ####?? How does this comment work?
    return gauss_points
end
```
