```julia
nx = 10;                     # so phan tu theo phuong x. (Luu y: nhap so chan) 
ny = 2; 
H  = 0.06;      # chieu cao
L  = 0.48;      # chieu dai


# Ma tran toa do x co kich thuoc (ny+1) * (nx+1)
# x = [x1 x2 ... xn;
#     ............;
#     x1 x2 ... xn]

# Ma tran toa do y co kich thuoc (ny+1) * (nx+1)
# y = [y1 y1 ... y1;
#      y2 y2 ... y2;
#      ............;
#      ym ym ... ym]

# Ma tran ung suat cung co kich thuoc (ny+1) * (nx+1)

x = linspace(0, L, nx+1) 
x = repmat(x',ny+1)

y = linspace(0, H, ny+1) 
y = repmat(y',nx+1)'


stress = rand(3,11)

@show x, y , stress
#Check the levels of stress
figure()
contourf(x,y,stress,  cmap=get_cmap("jet"))
interpolation = "bicubic"
colorbar()

figure()
l = linspace(0,1.05,50)
contourf(x,y,stress,  cmap=get_cmap("jet"), levels = l)
colorbar()

figure()
l = linspace(0,1.05,50)
contourf(x,y,stress,  cmap=get_cmap("gray"), levels = l)
colorbar()
```
